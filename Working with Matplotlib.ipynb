```python
# Another Dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# Create a histogram of age
heart_disease["age"].plot.hist();
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
## Matplotlib OO Method

#for quick plots plyplot method
# for advances use OO method
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
## OO Method
fig,ax = plt.subplots(figsize = (10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45,100])
```

```python
#OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))
# Plot the data
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],
                    c=over_50["target"]);
#Customize
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

#add Legend (using scatter variable created above)
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Subplot of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10));

# Add data to axis
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
ax0.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")
# Add Legend
ax0.legend(*scatter.legend_elements(),title ="Target");
# Add mean Line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# add data to axes for thalach
scatterthal= ax1.scatter(x=over_50["age"],
                        y=over_50["thalach"],
                        c=over_50["target"]);
ax1.set(title="Heart Disease and Thalach Levels",
       xlabel ="Age",
       ylabel ="ThanachLevels");
# add legend
ax1.legend(*scatterthal.legend_elements(),title="Target");
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle="--");

# Add title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight ="bold");
```

```python
#Set the style
plt.style.use('seaborn-whitegrid')

#OO method from scratch...with styling
fig,ax = plt.subplots(figsize=(10,6))
# Plot the data
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter");# this changes the color scheme
#Customize
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

#add Legend (using scatter variable created above)
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Customising the x and y axis
# Subplot of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10));

# Add data to axis
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap='winter');
ax0.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol");
ax0.set_xlim([50,80]);
# Add Legend
ax0.legend(*scatter.legend_elements(),title ="Target");
# Add mean Line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# add data to axes for thalach
scatterthal= ax1.scatter(x=over_50["age"],
                        y=over_50["thalach"],
                        c=over_50["target"],
                        cmap='winter');
ax1.set(title="Heart Disease and Thalach Levels",
       xlabel ="Age",
       ylabel ="ThanachLevels");
ax1.set_xlim([50,80]);
# add legend
ax1.legend(*scatterthal.legend_elements(),title="Target");
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle="--");

# Add title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight ="bold");
```
